```python
from turtle import title


pm25_df = pd.read_csv("./data/private-train/raw/input/S0000099-Phu Vien.csv")

for day in range(0, 21):
    values = []
    hours = [_ for _ in range(0, 24)]
    for h in range(0, 24):
        id = (day * 24) + h
        value_pm25 = pm25_df.iloc[id]['PM2.5']
        values.append(value_pm25)
    date = pm25_df.iloc[day * 24]['timestamp']
    plt.plot(hours, values)
    plt.title("Date: " + date)
    plt.show()
```
