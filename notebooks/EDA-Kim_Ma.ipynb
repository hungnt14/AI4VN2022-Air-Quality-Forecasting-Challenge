```python
def time_series(start, end, hour):
    time_series_df = df
    time_series_df =  time_series_df[['timestamp', 'PM2.5']][(
        time_series_df['timestamp'] >= start) & (time_series_df['timestamp'] <= end) & (time_series_df['timestamp'].dt.hour == hour)]

    tstmp = time_series_df['timestamp'][(
        time_series_df['timestamp'] >= start) & (time_series_df['timestamp'] <= end) & (time_series_df['timestamp'].dt.hour == hour)].unique()
    # time_series_df[['PM2.5', 'timestamp']].plot(figsize= (20, 10))
    # plt.show()
    # x = time_series_df['timestamp']
    # y = time_series_df['PM2.5']

    # plt.plot(x, y)
    # plt.xlabel('Time')
    # plt.ylabel('PM2.5 Value')
    # plt.title('PM2.5 Time Series')
    # return plt.show()

    fig, ax = plt.subplots(figsize= (15, 10))
    ax.plot(time_series_df.timestamp.unique(),
            time_series_df['PM2.5'], color='red', linewidth=1)
    ax.set_xlabel("Date")
    ax.set_ylabel("PM2.5")

    plt.show()
```

Visualize the distribution of PM2.5 value by month, hour

```python
def visualize_distribution_featue(feature):
    g = sns.FacetGrid(_df, col=feature, col_wrap=4)
    g.map(sns.histplot, 'PM2.5')
    plt.show()
```
